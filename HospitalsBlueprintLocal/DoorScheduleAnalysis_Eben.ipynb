Two functions below to clean the aggregated table with all doors exported from the ifc model. 

```python
for column in output_file:
    output_file[column] = output_file[column].astype(str).str.extract(r"\((.*)\)", expand=True)
    output_file[column] = output_file[column].astype(str).str.replace('\'','')
    
df_obj = output_file.select_dtypes(['string'])
output_file[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())  
#data = data.astype(str).split('(')[1].lstrip().rstrip('\)')
output_file.to_csv("fifthDraftDoors.csv")
```

Mapping function:

```python
def mapColumn(masterDictionary, masterCol, newColName, keY):
    
    dictLoc = dict(masterDictionary[[masterCol, newColName]].values)
    
    output_file["Mapped " + newColName] = output_file[keY].map(dictLoc)
```

```python
s_p_1.reset_index(drop=True, inplace=True)

def RepresentsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False

for i in range(0, len(s_p_1)):
    
    if RepresentsInt(s_p_1.loc[i, ['Ironmongery Set']]):
        s_p_1.loc[i,['Ironmongery Set']]=(int(s_p_1.loc[i,['Ironmongery Set']]))  
        
    if RepresentsInt(s_p_1.loc[i,["Mapped Ironmongery Set"]]):
        s_p_1.loc[i,['Mapped Ironmongery Set']]=(int(s_p_1.loc[i,['Mapped Ironmongery Set']]))
```
