```python
searchfor = ['Circulation', 'Ward', 'Plant and control', 'External']
df31 = df3[~df3['Department name'].str.contains('|'.join(searchfor))]
df31 = df31[df3.Area <= 80]

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize = (14,6))
sns.set_style('whitegrid')

# Violin plot
sns.violinplot(x='Department name', y='Area', data=df31)
plt.setp(ax.get_xticklabels(), rotation=40, horizontalalignment='right')
plt.title("ClatterBridge - Areas by Department")
plt.show()
```

```python
searchfor = ['Circulation', 'Ward', 'Plant and control', 'External']
df32 = df3[~df3['Department name'].str.contains('|'.join(searchfor))]
df32 = df32.groupby(['Department name'])['Area'].sum()

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize = (14,6))
plt.setp(ax.get_xticklabels(), rotation=40, horizontalalignment='right')
plt.title("ClatterBridge - Total Area by Department")

df32.plot.bar()
plt.show()
```

```python
#keys = pd.read_excel("Collated healtcare rooms schedule.xlsx", 'LOR Space Types', index_col = 1)
brighton_data = pd.read_excel("Collated healtcare rooms schedule_copy.xlsx", 'Brighton3Ts')
sccc_data = pd.read_excel("Collated healtcare rooms schedule_copy.xlsx", 'SCCC')
clatter = pd.read_excel("Collated healtcare rooms schedule_copy.xlsx", 'Clatterbridge')
dgri = pd.read_excel("Collated healtcare rooms schedule_copy.xlsx", 'DGRI')
data = pd.concat([brighton_data,sccc_data,clatter,dgri],axis=1)

#data = brighton_data.merge(sccc_data,left_on='LORSpaceType',right_on='LORSpaceType')
```

```python
space_types = [ 'Plant and control spaces',
               'Bedrooms', 'Medical spaces', 
               'Communication spaces', 'Storage spaces', 'Offices']

sccc_datav = sccc_data[sccc_data["LORSpaceType3"].isin(space_types)]
sccc_datav = sccc_datav[sccc_datav['SCCCArea(m2)'] <= 30]
clatter_datav = clatter[clatter["LORSpaceType2"].isin(space_types)]
clatter_datav = clatter_datav[clatter_datav['ClatterbridgeArea(m2)'] <=30]

plt.rcParams.update({'font.size': 11})

fig, ax = plt.subplots(figsize=(14, 10))
sns.set_style('whitegrid')
ax.set_xticks([], [])
ax.set_yticks([], [])
# Violin plots

ax = fig.add_subplot(211)
sns.violinplot(x=clatter_datav["LORSpaceType2"], y=clatter_datav["ClatterbridgeArea(m2)"], data= clatter_datav )
plt.title("Clatterbridge- Space Types")

#ax = fig.add_subplot(211)
#sns.violinplot(x=brighton_datav.LORSpaceType1, y=brighton_datav["BrightonArea(m2)"], data=brighton_datav, )
#plt.title("Brighton - Space Types")

ax = fig.add_subplot(212)
sns.violinplot(x=sccc_datav["LORSpaceType3"], y=sccc_datav["SCCCArea(m2)"], data=sccc_datav)
plt.title("SCCC - Space Types")
plt.show()
```

```python
space_types = ['Corridors', 'Plant and control spaces',
               'Bedrooms', 'Medical spaces', 
               'Communication spaces', 'Storage spaces']

brighton_datav = brighton_data[brighton_data["LORSpaceType1"].isin(space_types)]
brighton_datav = brighton_datav[brighton_datav["BrightonArea(m2)"] <= 200]


plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(figsize = (14,6))
sns.set_style('whitegrid')

# Violin plot
sns.violinplot(x=brighton_datav.LORSpaceType1, y=brighton_datav["BrightonArea(m2)"], data=brighton_datav)
plt.setp(ax.get_xticklabels(), rotation=40, horizontalalignment='right')
plt.title("Brighton - Common Space Types Under 200m2")
plt.show()
```
